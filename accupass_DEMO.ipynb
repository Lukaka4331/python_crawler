```python
#https://www.accupass.com/?area=north
from selenium import webdriver
ddriver = webdriver.Chrome()
#ddriver.get('https://www.accupass.com/?area=north')
ddriver.get('https://www.accupass.com/?area=north&channel=0')
```

```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
# çˆ¬å–ç¶²é å…§å®¹ï¼Œè§£æå¾Œèƒå–æ–°èæ‘˜è¦
import pandas as pd
infos_dict = {}
infos_dict['Time'] = []
infos_dict['Activity'] = []
infos_dict['Url'] = []
html = ddriver.page_source
soup = BeautifulSoup(html, "lxml")
all_kind = soup.find("div", attrs={'class':"style-f0d4a240-channel-container"}).find_all("div", attrs={'class':"style-bb526e47-theme-event-card-container"})

for news_item in all_kind:
    print("----------------------------------------------------------------------")
    news_time = news_item.find('span')    
    news_body = news_item.find('h3')
    print(news_time.text, news_body.text)
    #
    infos_dict['Time'].append(news_time.text)
    infos_dict['Activity'].append(news_body.text)
    
    # 
    # é€£åˆ°å¤–éƒ¨é€£çµï¼Œæ“·å–è©³ç´°æ–°èå…§å®¹
    #
    externalLink = "https://www.accupass.com" + news_item.a["href"]
    print(externalLink)
    #
    infos_dict['Url'].append(externalLink)
    
result_1=pd.DataFrame.from_dict(infos_dict)    
```
