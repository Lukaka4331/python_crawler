```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from tqdm.notebook import tqdm
```

```python
infos_dict = {}
infos_dict['Location'] = []#景點
data_dict ={}
data_dict['Location_URL'] = []#景點網站
```

```python
#Location
pagenum=63
pageurl = 'https://www.taiwan.net.tw/m1.aspx?sNo=0000064&page={}'
for page in tqdm(range(pagenum)):
    resp = requests.get(pageurl.format(page+1))
    soup = BeautifulSoup(resp.text, 'lxml')
    for tourist in soup.find_all("div",{"class":"card-title"}) :
        s1=tourist.text
        infos_dict['Location'].append(s1)
        print(s1)
    for tourist_url in soup.find("ul",{"class":"grid card-list card-style-columns"}).find_all("a",{"class":"card-link"}) :
        Url="https://www.taiwan.net.tw/" + tourist_url.get('href')
        data_dict['Location_URL'].append(Url)
        print(Url)
```

```python
infos_dict['recommend'] = []#讚數
a=['Oops!']
for i in tqdm(range(len(data_dict['Location_URL']))):
    resp = requests.get(data_dict['Location_URL'][i])
    soup = BeautifulSoup(resp.text, 'lxml')
    recommend_process=[recommend.text for recommend in soup.find("div",{"class":"sort-personal"}).find_all("b",{"class":"RecommendNumber"})]
    if recommend_process :
        infos_dict['recommend'].extend(recommend_process)
        print(i)
    else:
        print ("empty")
        infos_dict['recommend'].extend(a)
```

[pandas的cut，qcut函数的使用和区别](https://zhuanlan.zhihu.com/p/68194655)

```python
infos_dict['recommend_cut'] = []
infos_dict['recommend_cut']=infos_dict['recommend'].copy()
```

```python
for i in tqdm(range(len(infos_dict['Location']))): 
    infos_dict['recommend_cut'][i] = int(infos_dict['recommend_cut'][i]) 
```

```python
infos_dict['recommend_cut'] = pd.qcut(infos_dict['recommend_cut'], 3)
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
infos_dict['recommend_cut'] = label.fit_transform(infos_dict['recommend_cut'])
```

```python
result1=pd.DataFrame.from_dict(infos_dict)
```
