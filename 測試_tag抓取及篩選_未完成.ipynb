```python
res=requests.get(url)
soup= BeautifulSoup(res.text,'lxml')
```

```python
i=0
for tag in soup.find("div",{"class":"tagcollection"}).find_all("a",{"class":"tag-item"}):
    label_url='https://www.taiwan.net.tw/'+tag.get('href')
    data['Label'].append(label_url)
    label_text=tag.text
    label['list'].append(label_text)
    i=i+1    
    print(i)
```
