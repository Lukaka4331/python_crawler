# 政府開放平台_即時水位資料_api_抓取測試

資源:[即時水位資料](https://data.gov.tw/dataset/25768)

資源:[API](http://odata.wra.gov.tw/v4/RealtimeWaterLevel)

```
import requests
import json

URL = "http://odata.wra.gov.tw/v4/RealtimeWaterLevel" 
r = requests.get(URL)
json.loads(r.text)
```

主要欄位說明: 
* RecordTime(資料時間)
* StationIdentifier(站號)
* WaterLevel(水位)

```
import pandas as pd 

data = json.loads(r.text)
s1=pd.DataFrame(data['value'])
s1.head()
```

```
s1.to_csv('result.csv',index=False)
```
