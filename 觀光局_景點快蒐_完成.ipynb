# Location

```python
import pandas as pd
infos_dict = {}
infos_dict['Location'] = []
#infos_dict['phone'] = []
#infos_dict['address'] = []
#infos_dict['ip'] = []
#infos_dict['webisite'] = []
```

```python
infos_dict
```

```python
pagenum=63
pageurl = 'https://www.taiwan.net.tw/m1.aspx?sNo=0000064&page={}'
```

```python
for i in range(pagenum):
    resp = requests.get(pageurl.format(i+1))
    soup = BeautifulSoup(resp.text, 'lxml')
    for tag in soup.find_all("div",{"class":"card-title"}) :
        s1=tag.text
        infos_dict['Location'].append(s1)
        print(s1)
```

```python
result=pd.DataFrame.from_dict(infos_dict)
```

# 經緯度(ip)

```python
infos_dict['ip'] = []
```

```python
import re
```

```python
for i in range(len(index_list)):
    resp = requests.get(index_list[i])
    soup = BeautifulSoup(resp.text, 'lxml')
    #ip=[tag.text for tag in soup.find("dl",{"class":"info-table"}).find_all("dd",text=re.compile("12(0|1|2|3|4|5|6)+/"))]

    #if ip :
    #    infos_dict['ip'].extend(ip)
    #    print(i)
    #else:
    #    print ("empty")
    #    infos_dict['ip'].extend(a)
    ip=[tag.text for tag in soup.find("dl",{"class":"info-table"}).find_all("dd")]
    ip_process ="".join(ip)
    ip_final = re.findall("[0-9]+.[0-9]+/[0-9]+.[0-9]+", ip_process)    
    
    if ip_final :
        infos_dict['ip'].extend(ip_final)
        print(i)
    else:
        print ("empty")
        infos_dict['ip'].extend(a)    
```
