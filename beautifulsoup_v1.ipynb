```python
import requests
from bs4 import BeautifulSoup
response = requests.get("https://www.foodpanda.com.tw/en/city/taichung-city")
soup = BeautifulSoup(response.text, 'lxml')
```

找出一個( li )標籤

```python
print(soup.find("li"))
```

輸出文字中對應的標籤

```python
print(soup.find("span",{"class":"name fn"}))
```

輸出純文字

```python
print(soup.find("span",{"class":"name fn"}).text)
```

找出所有span標籤中對應的文字屬性

```python
print(soup.find_all("span",{"class":"name fn"}))
```

# 用法跟find一樣，但是回傳的是Python的list

```python
a=soup.find_all("span",{"class":"name fn"})
print(a)
```

用迴圈輸出字串的資料

```python
for tag in soup.find_all("span",{"class":"name fn"}) :
    print(tag.text)
```

```python
food = [tag.text for tag in soup.find_all("span",{"class":"name fn"})]
```

# BeautifulSoup + regular expression

找出漢堡關鍵字

```python
import re
print(soup.find_all("span",{"class":"name fn"}, text=re.compile("漢堡")))
```

```python
for tag in soup.find_all("span",{"class":"name fn"}, text=re.compile("漢堡")) :
    print(tag.text)
```
