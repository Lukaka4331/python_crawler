① 取出今天所有的發文

```python
from datetime import date
date_today = date.today()
date_today = str(date_today).replace('-','/')

import re
def get_date_post(text):
    regex = "\d+/\d+/\d+"
    pattern = re.compile(regex)
    result = re.search(pattern, text)
    date_post = result.group(0)
    return date_post

import pandas as pd

columns = ["Date", "Tag", "Title"]
df_infos = pd.DataFrame(columns=columns)
for d in soup.find(class_="part_list_2").find_all('h3')[1:]:
    date_post = get_date_post(d.find(class_="date").text)
    if date_post == date_today:
        tag = d.find("em").text
        title = d.find("a").text
        df_info = pd.DataFrame([[date_post, tag, title]], columns=columns)
        df_infos = df_infos.append(df_info, ignore_index=True)
    #break
df_infos
```

② 如果想要依照類別分類，怎麼儲存會比較好？

```python
df_infos.sort_values(by=['Tag']).reset_index().head(15)
```

③ 哪一個類別的文章最多

```python
pd.DataFrame.from_dict(dict(df_infos.groupby("Tag").size().sort_values(ascending=False)), orient='index', columns=["數量"])
```
