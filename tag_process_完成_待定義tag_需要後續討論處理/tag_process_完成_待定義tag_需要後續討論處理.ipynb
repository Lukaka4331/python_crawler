```python
import requests
from bs4 import BeautifulSoup
#import time
import pandas as pd
#import re
from tqdm.notebook import tqdm
infos_dict = {}
infos_dict['Location'] = []
data_dict ={}
data_dict['Location_URL'] = []
infos_dict['hash_tag'] = []
a=['#古蹟巡禮']
```

```python
#Location
pagenum=63
pageurl = 'https://www.taiwan.net.tw/m1.aspx?sNo=0000064&page={}'
for page in tqdm(range(pagenum)):
    resp = requests.get(pageurl.format(page+1))
    soup = BeautifulSoup(resp.text, 'lxml')
    for tourist in soup.find_all("div",{"class":"card-title"}) :
        s1=tourist.text
        infos_dict['Location'].append(s1)
        print(s1)
    for tourist_url in soup.find("ul",{"class":"grid card-list card-style-columns"}).find_all("a",{"class":"card-link"}) :
        Url="https://www.taiwan.net.tw/" + tourist_url.get('href')
        data_dict['Location_URL'].append(Url)
        print(Url)     
```

```python
for t_hash_tag in tqdm(range(len(data_dict['Location_URL'])),desc='hash_tag'):
            resp_1 = requests.get(data_dict['Location_URL'][t_hash_tag])
            soup_1 = BeautifulSoup(resp_1.text, 'lxml')
            hash_tag=[tag.text for tag in soup_1.find("div",{"class":"infoside"}).find_all("div",{"class":"hashtag"})]
            if hash_tag :
                infos_dict['hash_tag'].extend(hash_tag)
                print(t_hash_tag)
            else:
                print ("empty")#247-empty
                infos_dict['hash_tag'].extend(a)#tag訂為:古蹟巡禮        
```
