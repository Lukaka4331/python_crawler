```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
import pandas as pd
infos_dict = {}
infos_dict['Location'] = []
#infos_dict['phone'] = []
#infos_dict['address'] = []
#infos_dict['ip'] = []
#infos_dict['webisite'] = []
```

```python
infos_dict
```

```python
pagenum=63
pageurl = 'https://www.taiwan.net.tw/m1.aspx?sNo=0000064&page={}'
```

```python
for i in range(pagenum):
    resp = requests.get(pageurl.format(i+1))
    soup = BeautifulSoup(resp.text, 'lxml')
    for tag in soup.find_all("div",{"class":"card-title"}) :
        s1=tag.text
        infos_dict['Location'].append(s1)
        print(s1)
    
```

```python
result=pd.DataFrame.from_dict(infos_dict)
```

```python
import pandas as pd
result=pd.read_csv('page_all_final.csv')
```

```python
index_list=[]
for  name in result['Url']:
    index_list.append(name)
```

```python
import numpy as np
a=np.nan
```

```python
for i in range(len(index_list)):
    resp = requests.get(index_list[i])
    soup = BeautifulSoup(resp.text, 'lxml')
    phone=[tag.text for tag in soup.find("dl",{"class":"info-table"}).find_all("a",{"class":"tel-link phone"})]
    if phone :
        infos_dict['phone'].append(phone)
        print(i)
    else:
        print ("empty")
        infos_dict['phone'].append(a)
```

```python
for i in range(len(index_list)):
    resp = requests.get(index_list[i])
    soup = BeautifulSoup(resp.text, 'lxml')
    address=[tag.text for tag in soup.find("dl",{"class":"info-table"}).find_all("a",{"class":"tel-link address"})]
    if address :
        infos_dict['address'].append(address)
        print(i)
    else:
        print ("empty")
        infos_dict['address'].append(a)
```
