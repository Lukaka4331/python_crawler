```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
import pandas as pd
infos_dict = {}
infos_dict['Time'] = []
infos_dict['News'] = []

for demo in soup.find(class_="part_list_2").find_all('h3'):
    t1=demo.find(class_="date").text
    t2=demo.find_all('a')[-1].text
    infos_dict['Time'].append(t1)
    infos_dict['News'].append(t2) 
    
    print(demo.find(class_="date").text, demo.find_all('a')[-1].text)
    #print(demo.find(class_="date").text, demo.find_all('a')[0].text)
result_news=pd.DataFrame.from_dict(infos_dict)    
```
